# 1D Wave Equation Modeling Using PINNs
The goal of this notebook is to use a Physics-Informed Neural Net (PINN) to predict the seismic wavefield generated by some (given) initial conditions in a 1D acoustic isotropic medium with **constant velocity**. 

## Outline
- We create a neural net which takes into input the position $x$ and the time $t$. The output is the acoustic pressure $p(x,t)$ at this location/time
- We start with some initial conditions and we enforce a boundary condition on the 1D modeling domain: 
    1. Two initial conditions on the pressure field itself and on its time derivative
    $$ p(x,t=0) = sin(x) $$ and $$\frac{\partial p(x,t)}{\partial t}|_{t=0} = sin(x),$$ for all $x \in [0, L]$, where $L$ is the length of the domain.   
    2. One boundary condition 
    $$p(x,t) = 0,$$ for all $t$ and $x\in \{0,L\}$.
    3. One constraint enforcing the pressure field to satisfy the wave-equation inside the 1D modeling domain:
        $$\frac{1}{v^2}\frac{\partial^2 p(x,t)}{\partial t^2} - \frac{\partial^2 p(x,t)}{\partial x^2} = 0,$$ for all $x \in [0,L]$ and $t \in [0, 2L].$ 
- For the training stage of the PINN, we add these four constraints
- In this notebook, we compare the **PINN solution** with the **analytical solution** and the numercial solution obtained using a **finite-difference (FD)** scheme  

## Finite-difference solution

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
################## Analytical solution ##################
# Domain dimensions
L = float(np.pi)
x = np.arange(0, L, 0.01)
t = np.arange(0, 2*L, 0.01)
nx = x.shape[0]
nt = t.shape[0]
X, T = np.meshgrid(x, t)

# Analytical solution
u_a = np.zeros((nt, nx))
for it in range(t.shape[0]):
    for ix in range(x.shape[0]):
        u_a[it,ix] = np.sin(x[ix]) * (np.cos(t[it])+np.sin(t[it]))
```

```python
################## FD solution ##################
# Sampling
dx = 0.01
dt = 0.01

# Boundaries
L = float(np.pi)
x0 = 0
x1 = np.pi
t0 = 0
t1 = 2*np.pi
fmax = 1.0/(2*np.pi)
s = dt*dt/(dx*dx)

# Generate 1D mesh
x_fd = np.arange(x0, x1, dx)
t_fd = np.arange(t0, t1, dt)
nx_fd = x_fd.shape[0]
nt_fd = t_fd.shape[0]
v_fd = 1

# FD stability/dispersion
courant = v_fd*dt/dx
disp = v_fd/(v_fd*fmax)
print("Courant:" , courant)
print("Dispersion:" , disp)

## Finite-difference solution
u_fd = np.zeros((nt_fd, nx_fd))
u_fd[0,:] = np.sin(x_fd)
u_fd[1,:] = np.sin(x_fd)*(dt+1)
for it in range(2,nt_fd,1):
    for ix in range(2,nx_fd-1,1):
        u_fd[it,ix] = 2*(1-s*v_fd*v_fd)*u_fd[it-1,ix] - u_fd[it-2,ix] + s*v_fd*v_fd*(u_fd[it-1,ix+1]+u_fd[it-1,ix-1]) 
```

## PINN solution
Modify the file ``problem.py`` and train the PINN. Then, load the solution and compare the predicted pressure wavefield with the analytical and finite-difference solutions

```python
################## Load PINN solution ##################
import training.stage1.infer as infer_stage1

# Display information on inferencer
infer_stage1.info

# Predicted solution
u1 = infer_stage1.infer(x=X, t=T)['u']
u1 = u1.reshape(X.shape)
```

```python
################## Plotting results ##################
# We plot the predicted wavefield using three methods:
# 1. PINN
# 2. Analytical solution
# 3. Finite-difference solution

# Set figure size
plt.rcParams["figure.figsize"] = (18,5.5)

# Create figures and axes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Clipping
p_min = u1.min() 
p_max = u1.max() 

# PINN solution
img0 = axs[0].imshow(u_a, cmap='jet', vmin=p_min, vmax=p_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[0])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img0, cax=cax)
axs[0].set_xlabel('X', fontsize=18)
axs[0].set_ylabel('Time', fontsize=18)
axs[0].set_title('Pressure (Analytical)', fontsize=18)
axs[0].grid()
plt.tight_layout()

# Analytical solution
img1 = axs[1].imshow(u1, cmap='jet', vmin=p_min, vmax=p_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[1])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img1, cax=cax)
axs[1].set_xlabel('X', fontsize=18)
axs[1].set_ylabel('Time', fontsize=18)
axs[1].set_title('Pressure (PINN)', fontsize=18)
axs[1].grid()
plt.tight_layout()

# FD solution
img2 = axs[2].imshow(u_fd, cmap='jet', vmin=p_min, vmax=p_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[2])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img2, cax=cax)
axs[2].set_xlabel('X', fontsize=18)
axs[2].set_ylabel('Time', fontsize=18)
axs[2].set_title('Pressure (FD)', fontsize=18)
axs[2].grid()
plt.tight_layout()

# Save figure
plt.savefig("we-fwd-1d-cv-pressure.png")
```

```python
############### Plot error ###############
# Set figure size
plt.rcParams["figure.figsize"] = (18,5.5)

# Create figures and axes
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

fd_error = u_a - u_fd
pinn_error = u1 - u_fd
print("Max FD error: ", abs(fd_error).max())
print("Max PINN error: ", abs(pinn_error).max())

# Clipping
d_min = -0.02
d_max = 0.02

# FD error 
img0 = axs[0].imshow(fd_error, cmap='jet', vmin=p_min, vmax=p_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[0])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img0, cax=cax)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Time')
axs[0].set_title('FD error (high clipping)')
axs[0].grid()
plt.tight_layout()

# Analytical solution
img1 = axs[1].imshow(pinn_error, cmap='jet', vmin=p_min, vmax=p_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[1])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img1, cax=cax)
axs[1].set_xlabel('X')
axs[1].set_ylabel('Time')
axs[1].set_title('PINN error (high clipping)')
axs[1].grid()
plt.tight_layout()

# FD error 
img0 = axs[2].imshow(fd_error, cmap='jet', vmin=d_min, vmax=d_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[2])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img0, cax=cax)
axs[2].set_xlabel('X')
axs[2].set_ylabel('Time')
axs[2].set_title('FD error (low clipping)')
axs[2].grid()
plt.tight_layout()

# Analytical solution
img1 = axs[3].imshow(pinn_error, cmap='jet', vmin=d_min, vmax=d_max, extent=(0, L, 2*L, 0))
divider = make_axes_locatable(axs[3])
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(img1, cax=cax)
axs[3].set_xlabel('X')
axs[3].set_ylabel('Time')
axs[3].set_title('PINN error (low clipping)')
axs[3].grid()
plt.tight_layout()
```
